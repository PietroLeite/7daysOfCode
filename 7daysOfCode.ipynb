```python
def definir_classe(valor):
    if valor < 0 or valor > 999:
        return 'Classe fora do padrão'

    classes = {
        range(0, 100): 'Generalidades. Ciência e conhecimento',
        range(100, 200): 'Filosofia e psicologia',
        range(200, 300): 'Religião',
        range(300, 400): 'Ciências sociais',
        range(400, 500): 'Classe vaga',
        range(500, 600): 'Matemática e ciências naturais',
        range(600, 700): 'Ciências aplicadas',
        range(700, 800): 'Belas artes',
        range(800, 900): 'Linguagem. Língua. Linguística',
        range(900, 1000): 'Geografia. Biografia. História'
    }

    for intervalo, classe in classes.items():
        if valor in intervalo:
            return classe

    return 'Classe fora do padrão' 
```

```python
tabela_Emprestimo_completa['classe_geral'] = tabela_Emprestimo_completa['localizacao'].map(definir_classe)
```

<h3>Função para gerar o boxplot</h3>

```python
def gerar_boxplot(data,x,y,titulo):
    sns.set_theme(palette='Blues')
    grafico = sns.boxplot(y= y, x= x, data= data, color='cyan') 
    
     
    plt.xlabel(None)
    plt.ylabel(None)

    grafico.set_title(titulo,loc='left', weight='bold')
```

<h3>Calculando o diferença percentual</h3>

```python
tabela_pos_graduacao_desde_2017_pivotada.info()
```

```python
tabela_pos_graduacao_desde_2017_pivotada['2022'] = tabela_pos_graduacao_desde_2017_pivotada['2022'].astype('int')
```

```python
def diferenca_percentual(valor_posterior, valor_anterior):
    resultado = round(
        ((valor_posterior / valor_anterior) - 1) * 100
        ,2
    )
    return resultado
```

```python
percentual_2017_2018 = diferenca_percentual(tabela_pos_graduacao_desde_2017_pivotada.iloc[:,1],tabela_pos_graduacao_desde_2017_pivotada.iloc[:,0])
percentual_2018_2019 = diferenca_percentual(tabela_pos_graduacao_desde_2017_pivotada.iloc[:,2],tabela_pos_graduacao_desde_2017_pivotada.iloc[:,1])
percentual_2019_2022 = diferenca_percentual(tabela_pos_graduacao_desde_2017_pivotada.iloc[:,3],tabela_pos_graduacao_desde_2017_pivotada.iloc[:,2])
```

```python
percentual_2017_2018
```

```python
percentuais = pd.DataFrame({
    '2017 - 2018' : percentual_2017_2018,
    '2018 - 2019' : percentual_2018_2019,
    '2019 - 2022' : percentual_2019_2022
})
```

<h3>Formatando os dados da tabela</h3>

pedido da blibioteca:
* Não contenha numeração de índice;
* 
Os nomes dos cursos tenham apenas a primeira letra maiúscula
* Os números percentuais estejam indicados pelo símbolo “%”;
* Cor dos números: Positivos = Verde / Negativos = Vermelholho

```python
percentuais = percentuais.reset_index()
percentuais['CURSO'] = percentuais['CURSO'].str.capitalize()
percentuais.iloc[:, 1:4] = percentuais.iloc[:, 1:4].astype('string')
percentuais.iloc[:, 1:4] = percentuais.iloc[:, 1:4] + "%"
```

```python
percentuais
```

```python
#estilizando 
def colorir_valores(x):
    if "-" in x:
        cor = "red"
    elif "a" in x or "e" in x:
        cor = "white" 
    else:
        cor = "green" 
    return f'color: {cor}' 
percentuais_estilizado = percentuais.style.applymap(colorir_valores)
percentuais_estilizado
```

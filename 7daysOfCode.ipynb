```python
def definir_classe(valor):
    if valor < 0 or valor > 999:
        return 'Classe fora do padrão'

    classes = {
        range(0, 100): 'Generalidades. Ciência e conhecimento',
        range(100, 200): 'Filosofia e psicologia',
        range(200, 300): 'Religião',
        range(300, 400): 'Ciências sociais',
        range(400, 500): 'Classe vaga',
        range(500, 600): 'Matemática e ciências naturais',
        range(600, 700): 'Ciências aplicadas',
        range(700, 800): 'Belas artes',
        range(800, 900): 'Linguagem. Língua. Linguística',
        range(900, 1000): 'Geografia. Biografia. História'
    }

    for intervalo, classe in classes.items():
        if valor in intervalo:
            return classe

    return 'Classe fora do padrão' 
```

```python
tabela_Emprestimo_completa['classe_geral'] = tabela_Emprestimo_completa['localizacao'].map(definir_classe)
```

```python
data_emprestimos = pd.DataFrame(tabela_Emprestimo_completa['data_emprestimo'].value_counts()).reset_index()
```

```python
data_emprestimos.columns = ['data_do_emprestimo', 'quantidade']
```

```python
data_emprestimos['data_do_emprestimo'] = pd.to_datetime(data_emprestimos['data_do_emprestimo'])
data_emprestimos
```

```python
#agora podemos trabalhar melhor a coluna de tempo 
data_por_ano = pd.DataFrame(data_emprestimos.quantidade.groupby(data_emprestimos.data_do_emprestimo.dt.year).sum())
data_por_ano
```

```python
sns.set_theme(style="darkgrid", 
              font_scale=1.4, 
              context='notebook',
              rc= {"figure.figsize":(16,8)})
```

```python
grafico_linha = sns.lineplot(data=data_por_ano, x='data_do_emprestimo',y='quantidade')
grafico_linha.set(xlabel=None,ylabel=None)
grafico_linha.set_title('Quantidade de exemplares emprestados por ano', loc='left', weight='bold')
```

Com a analise do gráfico podemos notar um pico no ano de 2013 e nos anos posteriores as vendas vem decaindo

<h1>Observando por meses</h1>

```python
data_por_mes = pd.DataFrame(data_emprestimos.quantidade.groupby(data_emprestimos.data_do_emprestimo.dt.month).sum())
data_por_mes
```

<h3>Melhorando a observação da tabela</h3>

```python
data_por_mes.index.name = 'mes'
meses = {1:'Jan',
         2:'Fev',
         3:'Mar',
         4:'Abr',
         5:'Mai',
         6:'Jun',
         7:'Jul',
         8:'Ago',
         9:'Set',
         10:'Out',
         11:'Nov',
         12:'Dez'}
data_por_mes.index = data_por_mes.index.map(meses)
```

```python
data_por_mes
```

```python
grafico_linha = sns.lineplot(data=data_por_mes, x='mes',y='quantidade')
grafico_linha.set(xlabel=None,ylabel=None)
grafico_linha.set_title('Quantidade de exemplares emprestados por mes', loc='left', weight='bold')
```

Podemos visuzalizar que os meses de Março e Agosto são os meses com maior demanda.

E os meses de Julho e Dezembro são as com as menores demandas. 

Com isso podemos repassar a diretoria que bons meses para se aplicar a folga dos colaboradores podem ser os meses de Janeiro, Junho e Novembro onde o movimento da biblioteca começa a diminuir e nos periodos de pouco movimento tarefas importantes como o inventário possam ser realizadas com mais eficiência.

E como um alerta podemos averiguar com a diretoria sobre o motivo real da queda desses meses de menor movimentação. E assim discutir ideias como uma nova campanha de marketing voltada para esses periodos.

```python
data_por_hora = pd.DataFrame(data_emprestimos.quantidade.groupby(data_emprestimos.data_do_emprestimo.dt.hour).sum())
data_por_hora.index.name = 'horas'
data_por_hora = data_por_hora.reset_index()
data_por_hora
```

```python
grafico_barra = sns.barplot(data=data_por_hora, y='quantidade', x='horas', palette='Blues', dodge=False)
grafico_barra.set(xlabel=None,ylabel=None)
plt.legend([],[], frameon=False)
grafico_barra.set_title('Movimentação em relação a faixa horária', loc='left')
```

Com a observação do gráfico podemos dizer que os horários de menor movimentação sao os horarios entre as 6 as 9h e entre as 19 as 23h
Assim a diretoria pode atribuir aos colaboradores foco em outras atividades além do atendimento ao cliente
